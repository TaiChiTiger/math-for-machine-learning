### DEM

```python
elevation = np.loadtxt("G:/Data/GIS/raster/dem/dem.asc", skiprows=6)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(elevation, cmap='gray')
plt.show()
```

```python
x_max = elevation.shape[1]
y_max = elevation.shape[0]

x = np.arange(x_max)
y = np.arange(y_max)

X, Y = np.meshgrid(x, y) 
Z = elevation
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(18, 7))  
ax = fig.add_subplot(111, projection='3d')  
elevation_plot = ax.plot_surface(X, Y, Z, rstride=50, cstride=50, cmap='jet')
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$elevation$", fontsize=18)
ax.view_init(60, 140)
fig.colorbar(elevation_plot)
plt.show()
```

```python
fig = plt.figure(figsize=(18, 10))  
ax = fig.add_subplot(111, projection='3d')  

ax.plot_surface(X, Y, Z, rstride=50, cstride=50, alpha=0.3)
cset = ax.contour(X, Y, Z, zdir='z', offset=-100, cmap='coolwarm')
cset = ax.contour(X, Y, Z, zdir='x', offset=x.max()+300, cmap='coolwarm')
cset = ax.contour(X, Y, Z, zdir='y', offset=-100, cmap='coolwarm')
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$elevation$", fontsize=18)
ax.view_init(60, 140)
plt.show()
```

```python
fig = plt.figure(figsize=(12, 7))  
ax = fig.add_subplot(111)  
ax.contour(X, Y, Z, cmap='coolwarm')
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
plt.show()
```

```python
fig = plt.figure(figsize=(15, 7))  
ax = fig.add_subplot(111)  
contour_map = ax.contourf(X, Y , Z, cmap='rainbow')
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
fig.colorbar(contour_map)
plt.show()
```
