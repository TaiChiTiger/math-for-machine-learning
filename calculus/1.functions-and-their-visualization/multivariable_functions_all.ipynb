```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

# Scalar-based Function 

### 1. $f(x,y)=x+y$

```python
X = np.arange(-3, 3, 0.1)  
Y = np.arange(-3, 3, 0.1)  
X, Y = np.meshgrid(X, Y)    # x-y   
Z = X + Y

fig = plt.figure(figsize=(10,6))  
ax = fig.add_subplot(111, projection='3d')  
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))  
ax.set_title("$f(x,y)=x+y$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()
```

### $f(x,y)=x^2+y^2$

```python
X = np.arange(-3, 3, 0.1)  
Y = np.arange(-3, 3, 0.1)  
X, Y = np.meshgrid(X, Y) 
Z = (X ** 2 + Y ** 2)  

fig = plt.figure(figsize=(10,6))  
ax = fig.add_subplot(111, projection='3d')  
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))  
ax.set_title("$f(x,y)=x^2+y^2$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()  
```

```python
X = np.arange(-3, 3, 0.1)  
Y = np.arange(-3, 3, 0.1)  
X, Y = np.meshgrid(X, Y)
Z = (X ** 2 + Y ** 2)  

fig = plt.figure(figsize=(10,6))  
ax = fig.add_subplot(111)  
ax.contourf(X, Y , Z, cmap='rainbow')
ax.set_title("$f(x,y)=x^2+y^2$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", rotation=0, fontsize=18)
ax.set_xlim(-3.5, 3.5)
ax.set_ylim(-3.5, 3.5)
plt.show()  
```

### 2. $f(x,y)=sin(x+y)$

```python
X = np.arange(-np.pi, np.pi, 0.1)  
Y = np.arange(-np.pi, np.pi, 0.1)  
X, Y = np.meshgrid(X, Y)
Z = np.sin(X+Y)

fig = plt.figure(figsize=(10,6))  
ax = fig.add_subplot(111, projection='3d')  
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
ax.set_title("$f(x,y)=sin(x+y)$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()  
```

### 3. $f(x,y)=cos(x)sin(y)$

```python
X = np.arange(-np.pi, np.pi, 0.1)  
Y = np.arange(-np.pi, np.pi, 0.1)  
X, Y = np.meshgrid(X, Y)   
Z = np.sin(X) * np.cos(Y)

fig = plt.figure(figsize=(10,6))  
ax = fig.add_subplot(111, projection='3d')  
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))  
ax.set_title("$f(x,y)=cos(x)sin(y)$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()  
```

### 4. $f(x,y)=e^{(x+y)}$

```python
X = np.arange(-2, 2, 0.1)  
Y = np.arange(-2, 2, 0.1)  
X, Y = np.meshgrid(X, Y)    
Z = np.exp(X + Y)

fig = plt.figure(figsize=(10,6))  
ax = fig.add_subplot(111, projection='3d')  
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))  
ax.set_title("$f(x,y)=e^{(x+y)}$", fontsize=18)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
plt.show()  
```

# Vector-based Function 

## Parametric Curves (one input, two outputs)

### 1. $f(t)=\bigg[\begin{matrix} cos(t) \\ tsin(t)\end{matrix}\bigg]$

```python
t = np.arange(-np.pi, np.pi, 0.1)
X = np.cos(t)
Y = t * np.sin(t)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection="3d")
ax.set_title("$f(t)=[cos(t) \\ tsin(t)]$", fontsize=18)
ax.plot(X, Y, t, color='r')
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
ax.set_zlabel("$t$", fontsize=18)
ax.set_zlim(-2, 2)
plt.show()
```

Projected into two dimensional plane:

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)
t = np.arange(0, 4*np.pi, 0.1)
X = np.cos(t)
Y = t * np.sin(t)

ax.set_title("$f(t)=[cos(t) \ tsin(t)]$", fontsize=18)
ax.plot(X, Y, color='r')
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
ax.axis('equal')
plt.show()
```

### 2. $f(t) = \bigg[\begin{matrix} t cos(t) \\ t sin(t)\end{matrix}\bigg]$

```python
theta = np.arange(-np.pi, np.pi, 0.1)
X = theta * np.cos(theta)
Y = theta * np.sin(theta)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')  
ax.set_title("$f(t)=[tcos(t) , tsin(t)]$", fontsize=18)
ax.plot(X, Y, theta, color='r')
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
ax.set_zlabel("$$", fontsize=18)
plt.show()
```

Projected into two dimensional plane

```python
theta = np.arange(-2*np.pi, 2*np.pi, 0.1)
X = theta * np.cos(theta)
Y = theta * np.sin(theta)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.set_title("$f(t)=[tcos(t) , tsin(t)]$", fontsize=18)
ax.plot(X, Y, color='r')
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", rotation=0, fontsize=18)
ax.axis('equal')
plt.show()
```

## Parametric Surfaces (two inputs, three outputs)

### $f(t,s)=\bigg[\begin{matrix}3cos(t)+cost(t)cos(s) \\\    3sin(t)+sin(t)cos(s) \\\ sin(s)\end{matrix}\bigg]$

```python
t = np.arange(-np.pi, np.pi, 0.1)
s = np.arange(-np.pi, np.pi, 0.1)
t, s =np.meshgrid(t, s)
X = 3 * np.cos(t) + np.cos(t) * np.cos(s)
Y = 3 * np.sin(t) + np.sin(t) * np.cos(s)
Z = np.sin(s)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection="3d")
ax.set_title("$f(t,s)=[3cos(t)+cost(t)cos(s)$ , $3sin(t)+sin(t)cos(s)$ , $sin(s)]$", fontsize=18)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
ax.set_zlabel("$x_3$", fontsize=18)
ax.set_zlim(-2, 2)
plt.show()
```

## Vector Fields

### 1. $f(x,y)=\bigg[\begin{matrix} x \\ y \end{matrix}\bigg]$

```python
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
X1 = X
X2 = Y

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.quiver(X, Y, X1, X2, edgecolor='b', facecolor='None', linewidth=.5, cmap='autumn')
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", rotation=0, fontsize=18)
ax.set_title("$f(x,y)=[x \\ y]$", fontsize=18)
plt.show()
```

### 2. $f(x,y)=\bigg[\begin{matrix}y^3 - 9y \\ x^3-9x \end{matrix}\bigg]$ 

```python
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
X1 = Y**3 - 9*Y
X2 = X**3 - 9*X

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.quiver(X, Y, X1, X2, edgecolor='b', facecolor='None', linewidth=.5, cmap='autumn')
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", rotation=0, fontsize=18)
ax.set_title("$f(x,y)=[y^3 - 9y , x^3 - 9x]$", fontsize=18)
plt.show()
```

### 3. $f(x,y,z)=\bigg[\begin{matrix}x \\ y \\ z\end{matrix}\bigg]$

```python
X = np.arange(-5, 5, 1)
Y = np.arange(-5, 5, 1)
Z = np.arange(-5, 5, 1)
X, Y, Z = np.meshgrid(X, Y, Z)
X1 = X
X2 = Y
X3 = Z

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection="3d")
ax.quiver(X, Y, Z, X1, X2, X3, length=0.1)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
ax.set_title("$f(x,y,z)=[x , y , z]$", fontsize=18)
plt.show()
```

### 4. $f(x,y,z)=\bigg[\begin{matrix}yz \\ xz \\ xy\end{matrix}\bigg]$

```python
X = np.arange(-5, 5, 1)
Y = np.arange(-5, 5, 1)
Z = np.arange(-5, 5, 1)
X, Y, Z = np.meshgrid(X, Y, Z)
X1 = Y * Z
X2 = X * Z
X3 = X * Y

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection="3d")
ax.quiver(X, Y, Z, X1, X2, X3, length=0.05)
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$y$", fontsize=18)
ax.set_zlabel("$z$", fontsize=18)
ax.set_title("$f(x,y,z)=[yz , xz , xy]$", fontsize=18)
plt.show()
```

### $f(x,y)=\bigg[\begin{matrix} x^2+y^2 \\ x^2-y^2\end{matrix}\bigg]$

input space:

```python
X = np.arange(-3, 3, 0.1)  
Y = np.arange(-3, 3, 0.1)  
X, Y = np.meshgrid(X, Y)  

fig = plt.figure(figsize=(8, 8)) 
ax = fig.add_subplot(111)  

Xn = np.array([[0, 1, 0, 1, 3]])
Yn = np.array([[0, 0, 1, 1, 2]])
colors = ['r', 'k', 'b', 'g', 'y']
for i in range(Xn.shape[1]):
    ax.plot(Xn[:,i], Yn[:,i], marker='o', markersize=10, color=colors[i])
ax.set_xlim(-1, 19)
ax.set_ylim(-10, 10)
ax.grid(100)
plt.show()
```

output space:

```python
X = np.arange(-3, 3, 0.1)  
Y = np.arange(-3, 3, 0.1)  
X, Y = np.meshgrid(X, Y)   
X1 = X ** 2 + Y ** 2 
X2 = X ** 2 - Y ** 2 

fig = plt.figure(figsize=(8,8))  
ax = fig.add_subplot(111)  
ax.plot(X1, X2)  
ax.set_title("$f(x,y)=[x^2+y^2 \\ x^2-y^2]$", fontsize=18)
ax.set_xlim(-1, 19)
ax.set_ylim(-10, 10)

Xn = np.array([[0, 1, 0, 1, 3]])
Yn = np.array([[0, 0, 1, 1, 2]])
colors = ['r', 'k', 'b', 'g', 'y']
X1n = Xn ** 2 + Yn ** 2 
X2n = Xn ** 2 - Yn ** 2 
for i in range(Xn.shape[1]):
    ax.plot(X1n[:, i], X2n[:, i], marker='o', markersize=10, color=colors[i])
ax.grid()
plt.show()  
```
