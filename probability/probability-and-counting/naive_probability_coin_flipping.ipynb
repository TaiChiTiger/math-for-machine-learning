```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 500
np.random.seed(123)
outcomes = np.random.choice(a=(0, 1), p=(0.5, 0.5), size=N, replace=True)
```

```python
heads = np.cumsum(outcomes)
n = np.linspace(1, N, N)
run_prop = heads / n
run_prop[-1]
```

```python
plt.figure(figsize=(15, 5)) 
plt.plot(n, run_prop, '-o', )
plt.xscale('log')  
plt.xlim(1, N)
plt.ylim(0, 1)
plt.xlabel('Flip Number')
plt.ylabel('Proportion Heads')
plt.title('Running Proportion of Heads')
plt.axhline(y=.5, ls='dashed', c='red')

outcomes_seq = ''.join(["T","H"][outcome] for outcome in outcomes[:10])
plt.text(10, 0.8, 'Flip Sequence = {}...'.format(outcomes_seq))
plt.text(25, 0.2, 'End Proportion = {:2.3f}'.format(run_prop[-1]), 
         fontsize=14)
plt.show()
```
