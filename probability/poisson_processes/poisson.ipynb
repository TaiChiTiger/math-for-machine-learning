```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

## $x \sim Pois(\lambda=np)$ 
###  PMF: $pmf(k)=\frac {\lambda^k e^{-\lambda}} {k!}$

```python
lambda_ = [0.5, 1.5, 3, 8]
k = np.arange(0, max(lambda_) * 3)
plt.figure(figsize=(10, 8))
for lam in lambda_:
    y = stats.poisson(lam).pmf(k)
    plt.plot(k, y, 'o-', label="$\\lambda$ = {:2.1f}".format(lam))
plt.legend()
plt.xlabel("$k$", fontsize=14)
plt.ylabel("$pmf(k)$", fontsize=14)
plt.show()
```
