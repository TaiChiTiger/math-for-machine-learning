```python
from scipy.stats import multivariate_normal
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import seaborn
```

```python
mu = np.array([5, 7])
covariance = np.array([[6, 3], [3, 4]])
covariance
```

```python
x1 = np.linspace(0, 20, 100)
x2 = np.linspace(0, 20, 100)
X1, X2 = np.meshgrid(x1, x2)
X = np.dstack((X1, X2))

rv = multivariate_normal(mu, covariance)
P = rv.pdf(X)
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, P, cmap=cm.autumn)
plt.show()
```

```python
fig = plt.figure(figsize=(12, 8))
plt.contourf(X1, X2, P)
plt.show()
```

```python
covariance2 = np.array([[6, 0], [0, 4]])
covariance2

rv = multivariate_normal(mu, covariance2)
P2 = rv.pdf(X)
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, P2, cmap=cm.autumn)
plt.show()
```

```python
fig = plt.figure(figsize=(12, 8))
plt.contourf(X1, X2, P2)
plt.show()
```
