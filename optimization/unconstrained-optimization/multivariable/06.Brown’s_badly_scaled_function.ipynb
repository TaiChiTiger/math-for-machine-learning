```python
import ipyvolume.pylab as plt3d
import numpy as np

X = np.linspace(-200, 2*10**6, 1000)
Y = np.linspace(-200, 200, 1000)
X, Y = np.meshgrid(X, Y)
Z = (X - 10**6)**2 + (Y - 2*1e-6)**2 + (X * Y - 2)**2

plt3d.clear()   
plt3d.plot_surface(X, Z, Y)
plt3d.ylabel("z")
plt3d.zlabel("y")
plt3d.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
cs = plt.contour(X, Y, Z)
plt.clabel(cs)
plt.show()
```
