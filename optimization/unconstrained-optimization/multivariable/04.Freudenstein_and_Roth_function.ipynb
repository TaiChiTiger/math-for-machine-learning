```python
import ipyvolume.pylab as plt3d
import numpy as np

X = np.linspace(-10, 12, 100)
Y = np.linspace(-10, 12, 100)
X, Y = np.meshgrid(X, Y)
Z = (-13 + X + ((5 - Y) * Y - 2) * Y)**2 + \
    (-29 + X + ((Y + 1) * Y - 14) * Y)**2

plt3d.clear()   
plt3d.plot_surface(X, Z, Y)
plt3d.ylabel("z")
plt3d.zlabel("y")
plt3d.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
cs = plt.contour(X, Y, Z)
plt.clabel(cs)
plt.show()
```
