### $f(x,y)=(1-x)^2+100(y-x^2)^2$

```python
import numpy as np
import ipyvolume.pylab as plt3d

X = np.arange(-2, 2+ 0.05,  0.05)
Y = np.arange(-1.5, 3+ 0.05,  0.05)
X, Y = np.meshgrid(X, Y)
Z = (1.-X)**2 + 100.*((Y-X**2))**2

plt3d.clear()
plt3d.plot_surface(X, Z, Y, color='green')
plt3d.ylabel("z")
plt3d.zlabel("y")
plt3d.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
cs = plt.contour(X, Y, Z)
plt.clabel(cs)
plt.show()
```
