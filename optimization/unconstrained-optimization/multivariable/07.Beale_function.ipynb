```python
import ipyvolume.pylab as plt3d
import numpy as np

X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = (1.5 - X * (1 - Y))**2 + (2.25 - X * (1 - Y**2))**2 +\
    (2.625 - X * (1 - Y**3))**2

plt3d.clear()   
plt3d.plot_surface(X, Z, Y)
plt3d.ylabel("z")
plt3d.zlabel("y")
plt3d.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
cs = plt.contour(X, Y, Z)
plt.clabel(cs)
plt.show()
```
